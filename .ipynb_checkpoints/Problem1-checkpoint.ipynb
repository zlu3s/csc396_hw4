```python
import random
import torch
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm

# enable tqdm in pandas
tqdm.pandas()

# set to True to use the gpu (if there is one available)
use_gpu = True

# select device
device = torch.device('cuda' if use_gpu and torch.cuda.is_available() else 'cpu')
print(f'device: {device.type}')

# random seed
seed = 1234

# set random seed
if seed is not None:
    print(f'random seed: {seed}')
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
```

```python
file = "assignment4-dataset.txt"
small = "small.txt"

with open(file, 'r') as f:
    lines = f.read().splitlines()

df = pd.DataFrame(lines, columns=["Line"])
print(df) 
```

```python
from datasets import Dataset, DatasetDict

ds = DatasetDict()
ds['text'] = Dataset.from_pandas(df)
```

```python
from transformers import RobertaTokenizer, RobertaModel

# Transformer name
name = "roberta-base"
tokenizer = RobertaTokenizer.from_pretrained(name)
model = RobertaModel.from_pretrained(name)

def tokenize_function(example):
    return tokenizer(example["Line"], truncation=True)
    
tokenized_ds = ds['text'].map(tokenize_function, batched=True)

tokenized_ds.to_pandas()
```

```python
output = model(**tokenized_ds
```
